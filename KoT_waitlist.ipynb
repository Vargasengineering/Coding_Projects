```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://www.analogman.com/kotdelay.htm"
print(url)
page = urlopen(url)
print(page)
```

```python
html_bytes = page.read()
html = html_bytes.decode("utf-8")
html = BeautifulSoup(html, 'html.parser')
print(html.prettify())
```
